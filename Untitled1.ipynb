```python
tmax = ds.sel(time='1995')['tmax']
tmax
```

```python
tmax_avg = tmax.rolling(time=15, center=True).mean()
tmax_diff = tmax - tmax_avg

myLocationIndex = dict(x=200, y=500)

# convert to pandas, i.e. pull small DataArray to single node memory
df_tmax = pd.DataFrame(tmax.isel(**myLocationIndex).to_pandas())
df_avg = pd.DataFrame(tmax_avg.isel(**myLocationIndex).to_pandas())
df_diff = pd.DataFrame(tmax_diff.isel(**myLocationIndex).to_pandas())
```

```python
DELTA_TEMP_THRESHOLD = 1 # Celcius

df = pd.concat([df_tmax, df_avg, df_diff], axis=1)
df.columns = ['tmax', 'df_avg', 'diff']
df = df.reset_index()

df['hot'] = df['diff'] > DELTA_TEMP_THRESHOLD #Celcius
df['label'] = df['hot'].diff().ne(False).cumsum()
df = df.reset_index()
# e.g. for window size=15, drops 14 rows of nans from head and tail combined:
df = df.dropna(subset=['diff'])  
# select only the "hot" days (above moving average):
df = df[df['hot']]
df
```

```python
PERSISTED_FOR = 2 #days

hp = df.groupby('label').agg({'time':[np.min, np.max], 'tmax':np.max}).reset_index()
hp.columns = ['label', 't1', 't2', 'tmax']
hp['period'] = hp['t2'] - hp['t1'] + pd.Timedelta(1,'D')
hp['long'] = hp['period'] > pd.Timedelta(PERSISTED_FOR,'D')
hp = hp[hp['long']]
hp
```
