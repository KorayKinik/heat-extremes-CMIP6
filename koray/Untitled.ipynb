```python
import pystac
import fsspec
import xarray as xr

account_name = "daymeteuwest"
container_name = "daymet-zarr"

collection = pystac.Collection.from_file(
    "https://planetarycomputer.microsoft.com/api/stac/v1/collections/daymet-daily-na"
)
asset = collection.assets["zarr-https"]

store = fsspec.get_mapper(asset.href)
ds = xr.open_zarr(store, **asset.extra_fields["xarray:open_kwargs"])
ds
```

```python
# min_lon = -130
# max_lat = -120
# min_lat = 40
# max_lon = 50

# mask_lon = (ds.lon >= min_lon) & (ds.lon <= max_lon)
# mask_lat = (ds.lat >= min_lat) & (ds.lat <= max_lat)
# tmax = ds.where(mask_lon & mask_lat, drop=True).sel(time='2019-08-01')['tmax']
# x=slice(40,50), y=slice(-130, -120), 

# tmax = ds.sel(time='1995-07-16')['tmax']
tmax = ds.sel(time='1995')['tmax']

# tmax.nc (degrees C): Daily maximum 2-meter air temperature.
tmax
```

```python
tmax.sel(time='1995-07-16').plot()
```

```python
# For a given calendar date in a given ensemble member, the T2m anomalies are computed by 
# removing the climatological mean, defined as the 15-day running mean of T2m centered around 
# that date and averaged over all ensemble members. [...] daily T2m anomaly in the 
# 99th (1st) percentile and larger (smaller) than 3 K (-1 K) for a sequence of at least five consecutive days.
# https://agupubs.onlinelibrary.wiley.com/doi/10.1029/2019MS001958
```

```python
tmax_avg = tmax.rolling(time=15, center=True).mean()  #.dropna("time")
```

```python
tmax_diff = tmax - tmax_avg
```

```python
ax = tmax_diff.sel(time='1995-07-16').plot()
```

```python
# This is the map projection we want to plot *onto*
# https://daac.ornl.gov/DAYMET/guides/Daymet_Daily_V4.html
import cartopy.crs as ccrs
map_proj = ccrs.LambertConformal(central_longitude=-100, central_latitude=42.5)

tmax.sel(time='1995-07-16').plot(
    transform=ccrs.PlateCarree(), 
    col='time', col_wrap=1,
    subplot_kws={'projection': map_proj}
)
```
